```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Define the system of equations
def f(x, y):
    return x - y - x**2 - y**2

def g(x, y):
    return x + y - x**2 - y**2

# Find the stationary points
x, y = sp.symbols('x y')
f_sym = f(x, y)
g_sym = g(x, y)

# Solve for the stationary points
stationary_points = sp.solve([f_sym, g_sym], (x, y))
print(f"Stationary Points: {stationary_points}")

# Analyze the Jacobian matrix at the stationary points
J = sp.Matrix([[sp.diff(f_sym, x), sp.diff(f_sym, y)], [sp.diff(g_sym, x), sp.diff(g_sym, y)]])

# Classify each stationary point
for point in stationary_points:
    J_at_point = J.subs({x: point[0], y: point[1]})
    eigenvalues = J_at_point.eigenvals()
    print(f"Stationary Point: {point}")
    print(f"Jacobian Matrix at {point}:\n{J_at_point}")
    print(f"Eigenvalues: {eigenvalues}")
    if all(val.is_real for val in eigenvalues):
        if all(val > 0 for val in eigenvalues):
            print("Classification: Unstable Node")
        elif all(val < 0 for val in eigenvalues):
            print("Classification: Stable Node")
        else:
            print("Classification: Saddle Point")
    else:
        if all(val.is_real == False for val in eigenvalues):
            if all(val.as_real_imag()[0] < 0 for val in eigenvalues):
                print("Classification: Stable Focus (Spiral)")
            else:
                print("Classification: Unstable Focus (Spiral)")
        else:
            print("Classification: Complex Nature")
    print()

# Example: Plot the vector field
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = f(X, Y)
V = g(X, Y)

plt.figure(figsize=(10, 6))
plt.streamplot(X, Y, U, V, color='r')
plt.title('Vector Field')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms

# Generator Model
class Generator(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(True),
            nn.Linear(128, 256),
            nn.ReLU(True),
            nn.Linear(256, 512),
            nn.ReLU(True),
            nn.Linear(512, output_dim),
            nn.Tanh()
        )

    def forward(self, x):
        return self.model(x)

# Discriminator Model
class Discriminator(nn.Module):
    def __init__(self, input_dim):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)

# Hyperparameters
z_dim = 100
image_dim = 28*28
batch_size = 64
lr = 0.0002
epochs = 100

# Data Loader
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
train_loader = torch.utils.data.DataLoader(datasets.MNIST('.', train=True, transform=transform, download=True), batch_size=batch_size, shuffle=True)

# Initialize models and optimizers
G = Generator(z_dim, image_dim)
D = Discriminator(image_dim)
criterion = nn.BCELoss()
optimizer_G = optim.Adam(G.parameters(), lr=lr)
optimizer_D = optim.Adam(D.parameters(), lr=lr)

# Training loop
for epoch in range(epochs):
    for batch_idx, (real_data, _) in enumerate(train_loader):
        batch_size = real_data.size(0)
        real_data = real_data.view(batch_size, -1)

        # Train Discriminator
        z = torch.randn(batch_size, z_dim)
        fake_data = G(z)
        real_labels = torch.ones(batch_size, 1)
        fake_labels = torch.zeros(batch_size, 1)

        D_real = D(real_data)
        D_fake = D(fake_data.detach())

        loss_D_real = criterion(D_real, real_labels)
        loss_D_fake = criterion(D_fake, fake_labels)
        loss_D = loss_D_real + loss_D_fake

        optimizer_D.zero_grad()
        loss_D.backward()
        optimizer_D.step()

        # Train Generator
        D_fake = D(fake_data)
        loss_G = criterion(D_fake, real_labels)

        optimizer_G.zero_grad()
        loss_G.backward()
        optimizer_G.step()

    print(f"Epoch [{epoch+1}/{epochs}] Loss D: {loss_D.item()}, loss G: {loss_G.item()}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import contextlib
import logging
import random 

# Generate a random time series dataset
np.random.seed(42)
n = 200
date_rng = pd.date_range(start='1/1/2000', periods=n, freq='M')
data = pd.Series(np.random.randn(n).cumsum(), index=date_rng)

# Visualize the data
data.plot(title='Randomly Generated Time Series')
plt.show()

# Check for stationarity
result = adfuller(data)
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Differencing if necessary
data_diff = data.diff().dropna()

# Plot ACF and PACF
plot_acf(data_diff, title='ACF of Differenced Data')
plot_pacf(data_diff, title='PACF of Differenced Data')
plt.show()

# Parameter grid search
#p = d = q = range(0, 3)
#pdq = [(x, y, z) for x in p for y in d for z in q]

# Seasonal parameter grid search
#seasonal_pdq = [(x, y, z, 12) for x in p for y in d for z in q]
# Define the range for p, d, q
p = d = q = range(0, 3)

# Generate all unique (p, d, q) combinations
all_pdq = [(x, y, z) for x in p for y in d for z in q]

# Sample 30 unique (p, d, q) combinations
pdq = random.sample(all_pdq, 3)

# Generate all unique seasonal (p, d, q, 12) combinations
all_seasonal_pdq = [(x, y, z, 12) for x in p for y in d for z in q]

# Sample 30 unique seasonal (p, d, q, 12) combinations
seasonal_pdq = random.sample(all_seasonal_pdq, 3)
print(len(pdq), len(seasonal_pdq))
best_aic = np.inf
best_pdq = None
best_seasonal_pdq = None

# Function to suppress logging output
@contextlib.contextmanager
def suppress_logging():
    logger = logging.getLogger('statsmodels')
    level = logger.getEffectiveLevel()
    logger.setLevel(logging.CRITICAL)
    try:
        yield
    finally:
        logger.setLevel(level)

# Grid search for best parameters
for i, param in enumerate(pdq):
    for seasonal_param in seasonal_pdq:
        print(i + 1, param, seasonal_param)
        with suppress_logging():
            mod = SARIMAX(data, order=param, seasonal_order=seasonal_param)
            results = mod.fit(disp=False)
            
            if results.aic < best_aic:
                best_aic = results.aic
                best_pdq = param
                best_seasonal_pdq = seasonal_param

print(f'Best SARIMAX order: {best_pdq} Seasonal order: {best_seasonal_pdq} AIC: {best_aic}')

# Fit the best model
model = SARIMAX(data, order=best_pdq, seasonal_order=best_seasonal_pdq)
results = model.fit(disp=False)

# Model diagnostics
results.plot_diagnostics(figsize=(15, 12))
plt.show()

# Display the summary of the model
print(results.summary())
```

```python
# Define the number of steps to forecast
n_steps = 15

# Get forecasted values
forecast = results.get_forecast(steps=n_steps)

# Get confidence intervals of the forecasts
forecast_ci = forecast.conf_int()

# Extract the predicted mean
predicted_mean = forecast.predicted_mean

# Print the forecasted values and confidence intervals
print(predicted_mean)
print(forecast_ci)

# Plot the historical data
plt.plot(data.index, data, label='Observed')

# Plot the forecasted values
forecast_index = pd.date_range(start=data.index[-1], periods=n_steps+1, freq='M')[1:]
plt.plot(forecast_index, predicted_mean, label='Forecast')

# Plot the confidence intervals
plt.fill_between(forecast_index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink', alpha=0.5)

plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.linalg import qr

# Define the Lorenz system
def lorenz(X, t, sigma, rho, beta):
    x, y, z = X
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# Parameters
sigma = 10
rho = 28
beta = 8/3
X0 = [1, 1, 1]  # Initial condition
t = np.linspace(0, 50, 10000)  # Time points

# Solve ODE
solution = odeint(lorenz, X0, t, args=(sigma, rho, beta))
x, y, z = solution.T

# Plot the Lorenz attractor
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

# Function to compute the Lyapunov exponent using QR decomposition
def lyapunov_exponent(lorenz, X0, t, sigma, rho, beta):
    dt = t[1] - t[0]
    X = np.array(X0)
    N = len(t)
    
    # Initial perturbation vectors
    e = np.eye(len(X0))
    Q = e.copy()
    L = np.zeros(len(X0))
    
    for i in range(N-1):
        X = odeint(lorenz, X, [t[i], t[i+1]], args=(sigma, rho, beta))[-1]
        Y = X + Q * 1e-8
        for j in range(len(X0)):
            Y[j] = odeint(lorenz, Y[j], [t[i], t[i+1]], args=(sigma, rho, beta))[-1]
        
        # QR decomposition
        Q, R = qr(Y - X)
        L += np.log(np.abs(np.diag(R)))
    
    return L / (N * dt)

# Calculate the Lyapunov exponent
lyap_exp = lyapunov_exponent(lorenz, X0, t, sigma, rho, beta)
print(f"Lyapunov Exponents: {lyap_exp}")

# Additional plotting for Lyapunov exponent over time
lyapunov_exp_list = []
delta0 = 1e-8
X1 = np.array(X0)
X2 = X1 + np.array([delta0, delta0, delta0])
delta = delta0

for i in range(len(t)-1):
    dt = t[i+1] - t[i]
    X1 = odeint(lorenz, X1, [t[i], t[i+1]], args=(sigma, rho, beta))[-1]
    X2 = odeint(lorenz, X2, [t[i], t[i+1]], args=(sigma, rho, beta))[-1]
    delta = np.linalg.norm(X2 - X1)
    lyapunov_exp_list.append(np.log(delta / delta0))
    delta0 = delta
    X2 = X1 + delta0 * (X2 - X1) / delta  # Re-normalize the perturbation vector

# Plotting the Lyapunov exponent over time
plt.figure(figsize=(10, 6))
plt.plot(t[:-1], np.cumsum(lyapunov_exp_list) / (t[:-1]+1e-10), label='Lyapunov Exponent')
plt.axhline(y=np.mean(lyap_exp), color='r', linestyle='--', label='Average Lyapunov Exponent')
plt.xlabel('Time')
plt.ylabel('Lyapunov Exponent')
plt.legend()
plt.title('Lyapunov Exponent over Time for the Lorenz System')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(r, x):
    return r * x * (1 - x)

def lyapunov_exponent(r, x0, n_iter):
    x = x0
    lyapunov_sum = 0
    
    for _ in range(n_iter):
        x = logistic_map(r, x)
        lyapunov_sum += np.log(abs(r - 2 * r * x))
    
    return lyapunov_sum / n_iter

# Parameters
r = 3.9  # Growth rate
x0 = 0.5  # Initial condition
n_iter = 10000  # Number of iterations

# Calculate Lyapunov exponent
lyapunov_exp = lyapunov_exponent(r, x0, n_iter)
print(f"Lyapunov Exponent for r={r}: {lyapunov_exp}")

# Plot Lyapunov exponent over a range of r values
r_values = np.linspace(2.5, 4.0, 1000)
lyapunov_exponents = [lyapunov_exponent(r, x0, n_iter) for r in r_values]

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(r_values, lyapunov_exponents, label='Lyapunov Exponent')
plt.axhline(0, color='red', linestyle='--')
plt.title('Lyapunov Exponent for Logistic Map')
plt.xlabel('r')
plt.ylabel('Lyapunov Exponent')
plt.legend()

# Visualize Logistic Map: Bifurcation Diagram
num_iterations = 1000
last = 100
x = 1e-5 * np.ones(1000)

# Initialize array to store results
results = []

# Iterate over values of r
for i in range(num_iterations):
    x = logistic_map(r_values, x)
    if i >= (num_iterations - last):
        results.append(x.copy())

# Convert results to a 2D array
results = np.array(results).T

# Plot bifurcation diagram
plt.subplot(1, 2, 2)
plt.plot(r_values, results, ',k', alpha=0.25)
plt.xlim(2.5, 4.0)
plt.title('Bifurcation Diagram of the Logistic Map')
plt.xlabel('r')
plt.ylabel('x')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ising_metropolis(N, J, T, steps):
    # Initialize spins
    spins = np.random.choice([-1, 1], size=(N, N))
    
    def energy(spins):
        return -J * np.sum(spins * (np.roll(spins, 1, axis=0) + np.roll(spins, -1, axis=0) + np.roll(spins, 1, axis=1) + np.roll(spins, -1, axis=1)))
    
    for step in range(steps):
        i, j = np.random.randint(0, N, 2)
        dE = 2 * J * spins[i, j] * (spins[i, (j-1)%N] + spins[i, (j+1)%N] + spins[(i-1)%N, j] + spins[(i+1)%N, j])
        if dE < 0 or np.random.rand() < np.exp(-dE / T):
            spins[i, j] *= -1
    
    return spins

# Parameters
N = 20  # Lattice size
J = 1   # Interaction energy
T = 2.0 # Temperature
steps = 1000 # Number of Monte Carlo steps

# Run Metropolis algorithm
spins = ising_metropolis(N, J, T, steps)

# Plot final spin configuration
plt.imshow(spins, cmap='coolwarm')
plt.colorbar(label='Spin')
plt.title('Ising Model Spin Configuration')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Uniform random numbers
uniform_random_numbers = np.random.uniform(0, 1, 10000)

# Inverse transform sampling for exponential distribution (lambda=1)
exponential_random_numbers = -np.log(1 - uniform_random_numbers)

# Box-Muller transform for normal distribution
U1 = np.random.uniform(0, 1, 5000)
U2 = np.random.uniform(0, 1, 5000)
Z0 = np.sqrt(-2 * np.log(U1)) * np.cos(2 * np.pi* U2)
Z1 = np.sqrt(-2 * np.log(U1)) * np.sin(2 * np.pi *U2)

# Plotting the histograms
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(uniform_random_numbers, bins=50, density=True)
plt.title('Uniform Distribution')

plt.subplot(2, 2, 2)
plt.hist(exponential_random_numbers, bins=50, density=True)
plt.title('Exponential Distribution')

plt.subplot(2, 2, 3)
plt.hist(Z0, bins=50, density=True, alpha=0.7, label='Z0')
plt.hist(Z1, bins=50, density=True, alpha=0.7, label='Z1')
plt.title('Normal Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def energy(state):
    return -np.sum(state * np.roll(state, 1))

def metropolis_step(state, beta):
    N = len(state)
    i = np.random.randint(0, N)
    state[i] *= -1
    dE = 2 * state[i] * (state[(i-1)%N] + state[(i+1)%N])
    if dE <= 0 or np.random.rand() < np.exp(-beta * dE):
        return state
    state[i] *= -1
    return state

# Parameters
N = 100  # Number of spins
beta = 1.0  # Inverse temperature
steps = 10000  # Number of Monte Carlo steps

# Initial state
state = np.ones(N)

# Run Metropolis algorithm
for step in range(steps):
    state = metropolis_step(state, beta)

# Plot final state
plt.plot(state, 'o-')
plt.title('Final State of Spins')
plt.xlabel('Spin Index')
plt.ylabel('Spin Value')
plt.show()
```
